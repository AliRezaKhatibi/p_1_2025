```python
#from graphviz import Digraph
import graphviz
input_nodes = 2
hidden_layers = [12,6]
output_nodes = 1



def visualize_neural_network(input_nodes, hidden_layers, output_nodes):
    # ساخت گراف
    dot = graphviz.Digraph(format="png")
    dot.attr(rankdir="LR")  # جهت گراف: چپ به راست
    
    # لایه ورودی
    for i in range(1, input_nodes + 1):
        dot.node(f'I{i}', f'Input {i}', shape='circle', style='filled', color='lightblue')
    
    # لایه‌های پنهان
    for layer_idx, num_nodes in enumerate(hidden_layers, start=1):
        for i in range(1, num_nodes + 1):
            dot.node(f'H{layer_idx}_{i}', f'H{layer_idx}.{i}', shape='circle', style='filled', color='lightgreen')
    
    # لایه خروجی
    for i in range(1, output_nodes + 1):
        dot.node(f'O{i}', f'Output {i}', shape='circle', style='filled', color='lightcoral')
    
    # اتصال لایه‌ها
    # از ورودی به اولین لایه پنهان
    for i in range(1, input_nodes + 1):
        for j in range(1, hidden_layers[0] + 1):
            dot.edge(f'I{i}', f'H1_{j}')
    
    # از لایه‌های پنهان به یکدیگر
    for layer_idx in range(1, len(hidden_layers)):
        for i in range(1, hidden_layers[layer_idx - 1] + 1):
            for j in range(1, hidden_layers[layer_idx] + 1):
                dot.edge(f'H{layer_idx}_{i}', f'H{layer_idx + 1}_{j}')
    
    # از آخرین لایه پنهان به خروجی
    for i in range(1, hidden_layers[-1] + 1):
        for j in range(1, output_nodes + 1):
            dot.edge(f'H{len(hidden_layers)}_{i}', f'O{j}')
    
    # نمایش گراف
    dot.attr(size='6,6')  # اندازه نمودار 6x6 اینچ
    dot.attr(ratio='1')   # نسبت تصویر مربعی
    dot.attr(dpi='600') 
    dot.render("neural_network", view=True)

# تنظیمات شبکه عصبی
#input_nodes          # تعداد نودهای ورودی
#hidden_layers    # تعداد نودها در هر لایه پنهان
#output_nodes          # تعداد نودهای خروجی

# فراخوانی تابع
visualize_neural_network(input_nodes, hidden_layers, output_nodes)
```
